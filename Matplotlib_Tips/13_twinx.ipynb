# Matplotlib: Secondary Axis with `twinx`

Import matplotlib's pyplot, numpy, and seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Create data

```python
x = np.linspace(0, 100, 1000)
x_sqrt = np.sqrt(x)
x_squared = x**2
```

## Basics

```python
plt.plot(x, x)
plt.plot(x, x_sqrt);
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x, x)
ax1.set_ylabel('Linear')

ax2 = ax1.twinx()
ax2.plot(x, x_sqrt, color='orange')
ax2.set_ylabel('Square Root');  #without orange, #add labels
```

```python
ax1.twinx() #open docs, then type()
```

## Tips

1. **Color code** the axes labels and tick labels (and spines)

```python
fig, ax1 = plt.subplots()

color1 = 'blue'
ax1.plot(x, x, color=color1)
ax1.set_ylabel('Linear', color=color1)
ax1.tick_params(axis='y', labelcolor=color1)
ax1.spines['left'].set_color(color1)

ax2 = ax1.twinx()

color2 = 'crimson'
ax2.plot(x, x_sqrt, color=color2)
ax2.set_ylabel('Square Root', color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

#color of spines
ax2.spines['left'].set_color(color1)
ax2.spines['right'].set_color(color2);
```

2. Add a **legend** if necessary

```python
fig, ax1 = plt.subplots()
line1 = ax1.plot(x, x, label='Linear')
ax1.set_ylabel('Linear')
ax1.legend()

ax2 = ax1.twinx()
line2 = ax2.plot(x, x_sqrt, color='orange', label='Sqrt')  
ax2.set_ylabel('Square Root')
ax2.legend(loc=1);
```

```python
fig, ax1 = plt.subplots()
line1 = ax1.plot(x, x, label='Linear')
ax1.set_ylabel('Linear')

ax2 = ax1.twinx()
line2 = ax2.plot(x, x_sqrt, color='orange', label='Sqrt')  
ax2.set_ylabel('Square Root')

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc=0);
```

3. Typically x-axis matches, y-axes have different scales -- **don't try to lie!**

```python
fig, ax1 = plt.subplots()
ax1.plot(x, x)
ax1.set_ylabel('Linear')

ax2 = ax1.twinx()
ax2.plot(x_sqrt, x_sqrt, color='orange')  #x to x_sqrt
ax2.set_ylabel('Square Root');
```

```python
plt.plot(x, x)
plt.plot(x_sqrt, x_sqrt);
```

4. Seaborn also allows `twinx`

```python
sns.set_style('darkgrid')

ax1 = sns.lineplot(x=x, y=x, ci=None)

ax2 = ax1.twinx()

sns.lineplot(x=x, y=x_sqrt, ci=None, ax=ax2, color='orange');
```

## BONUS Content

### `twiny`

```python
fig, ax = plt.subplots()

color1 = 'forestgreen'
ax.plot(x, x, color=color1, lw=3)
ax.set_xlabel('Linear', color=color1, fontsize=14)
ax.tick_params(axis='x', labelcolor=color1)

ax_new = ax.twiny()

color2 = 'deeppink'
ax_new.plot(x_sqrt, x, color=color2, lw=3)
ax_new.set_xlabel('Squared', color=color2, fontsize=14)
ax_new.tick_params(axis='x', labelcolor=color2)

ax_new.spines['bottom'].set_color('forestgreen')
ax_new.spines['top'].set_color('deeppink')
```

### Three Spines

```python
fig, ax1 = plt.subplots()

color1 = 'navy'
ax1.plot(x, x, color=color1, lw=3)
ax1.set_ylabel('Linear', color=color1, fontweight='bold', fontsize=14)
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
ax3 = ax1.twinx()

#Move third axis to the right to avoid overlap with second axis
ax3.spines['right'].set_position(("axes", 1.2))

color2 = 'crimson'
ax2.plot(x, x_sqrt, color=color2, lw=3)
ax2.set_ylabel('Square Root', color=color2, fontweight='bold', fontsize=14)
ax2.tick_params(axis='y', labelcolor=color2)

color3 = 'dodgerblue'
ax3.plot(x, x_squared, color=color3, lw=3)
ax3.set_ylabel('Squared', color=color3, fontweight='bold', fontsize=14)
ax3.tick_params(axis='y', labelcolor=color3)
```

## Overlay and Thumbnail

```python
sns.set_style('darkgrid')

fig, ax1 = plt.subplots()

color1 = 'navy'
ax1.plot(x, x, color=color1, lw=3)
ax1.set_ylabel('Linear', color=color1, fontweight='bold', fontsize=14)
ax1.tick_params(axis='y', labelcolor=color1, labelsize=12)

#plt.savefig('13twinx_linear.png', dpi=200)

ax2 = ax1.twinx()

color2 = 'crimson'
ax2.plot(x, x_sqrt, color=color2, lw=3)
ax2.set_ylabel('Square Root', color=color2, fontweight='bold', fontsize=14)
ax2.tick_params(axis='y', labelcolor=color2, labelsize=12)
ax2.grid(False)

#plt.savefig('13twinx_linear_with_secondary.png', dpi=200)
```
